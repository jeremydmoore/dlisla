```python
# OCR each image and output as PDF

pdf_directory_path = tif_directory_path.joinpath('_individual_pdfs')
pdf_directory_path.mkdir(exist_ok=True)

print(f'OCR {number_of_tifs} *.tif in {tiff_directory_path}')

# progress bar
progress_label = Label('OCR *.tif')
progress_bar = IntProgress(min=0, max=number_of_tifs)
progress_widget = VBox([progress_label, progress_bar])
display(progress_widget)

number_of_pdfs = 0

for image_path in image_paths_list:
    
    label = f'OCRing {image_path.name}'
    progress_label.value = label
    
    pdf_output_path = pdf_directory_path.joinpath(image_path.stem)
    
    # ocr *.tif with tesseract
    !tesseract {str(image_path)} {str(pdf_output_path)} pdf
    
    new_pdf_path = pdf_directory_path.joinpath(f'{image_path.stem}.pdf')  # tesseract doesn't want .pdf so we need to add it now
    
    if new_pdf_path.is_file():
        number_of_pdfs += 1
        progress_bar.value = number_of_pdfs
        
if number_of_pdfs != number_of_tifs:
    print(f'ERROR - ERROR - ERROR')
    print(f'# of *.tif: {number_of_tifs}')
    print(f'# of PDFs: {number_of_pdfs}')
else:
    print(f'{number_of_pdfs} PDFs created and saved in {pdf_directory_path}')
```

```python
# concatenate all *.pdf in local temp directory into single PDF

print(f'Combining *.pdf into single PDF')

combined_pdf_name = f'{tif_directory_path.stem}.pdf'
combined_pdf_save_path = tif_directory_path.joinpath(combined_pdf_name)
!pdftk {str(pdf_directory_path)}/*.pdf cat output {str(combined_pdf_save_path)}

if combined_pdf_save_path.is_file() and combined_pdf_save_path.stat().st_size > 0:
    print(f'{combined_pdf_save_path} created with size {round((combined_pdf_save_path.stat().st_size/1024/1024/1024), 2)} GB')
```
